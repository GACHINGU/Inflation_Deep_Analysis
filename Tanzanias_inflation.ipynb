```python
import requests
import pandas as pd

# --- Fetch Data ---
url = "https://api.worldbank.org/v2/country/TZA/indicator/FP.CPI.TOTL.ZG"
params = {"format": "json", "per_page": 1000}

response = requests.get(url, params=params)
data = response.json()

# --- Extract & Build DataFrame ---
df = pd.DataFrame(data[1])[["date", "value"]]

# --- Rename Columns (lowercase) ---
df = df.rename(columns={
    "date": "year",
    "value": "inflation"
})

# --- Clean Data ---
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["inflation"] = pd.to_numeric(df["inflation"], errors="coerce")

# Drop missing values
df = df.dropna()

# Convert year to integer
df["year"] = df["year"].astype(int)

# Sort by year
df = df.sort_values(by="year").reset_index(drop=True)

# Round inflation values
df["inflation"] = df["inflation"].round(2)

# --- Output ---
print(df.head())
print(df.tail())

# Save clean dataset
df.to_csv("tanzania_inflation_clean.csv", index=False)
```
